# LOGIN/SIGN UP PAGE

# LIBRARY MANAGEMENT PAGE

```python
print("\n We have the following types of Books:\n ",Book_list.keys())

#list of the actions that can be performed
#Add Book Details - Adding book to the list
#Delete Book - Deleting book from the list
#View Book List - To view the list of the books present
#Issue Book to Student - If the student wants to issue the book
#Return Book - When the student returns the book
#Exit - This will exit from the Library management book
Menu=["Add Book Details","Delete Book","View Book List","Issue Book to Student","Return Book","Exit"]

#when the user has logged in
while user==True:
    print("\n Input Actions are: \n",Menu)
    #input action that user wants to perform
    user_menu=input("\n Please input your action from the above list: \n")
    if user_menu==Menu[0]: #if user enters Add Book Details
        #input the type of book that user want to add to list
        Book_selected=input('\n Please enter which type of book you want to add to list \n Available options are "Fiction","Horror","Python","Maths": ')
        print("Books already in",Book_selected,"are: ",Book_list[Book_selected])
        #if the user inputs "Fiction" or "Horror" or "Python" or "Maths" system will ask for the bookid that needs to be added
        if Book_selected=="Fiction" or Book_selected=="Horror" or Book_selected=="Python" or Book_selected=="Maths":
            bookid=int(input("\n Please enter the book Id you want to add: "))
            title=input("Enter book title: ") #asking for the title of book
            status=input("Enter the status of book: ") #putting status of the book
            #the book will be added to the booklist of respective type and the bookid status will be added to the staus_id dictionary
            #when book selected is fiction
            if Book_selected=="Fiction" and bookid not in Book_list["Fiction"]:
                Book_list["Fiction"][bookid]=[title,status] #booklist will be added to fiction with its book id,title and status
                status_id[bookid]="available" #bookid will be added to status_id dict with {bookid:status}
                print("\n BOOK ID",bookid,"is added to the list \n")
            #when book selected is "Horror"
            elif Book_selected=="Horror" and bookid not in Book_list["Horror"]:
                Book_list["Horror"][bookid]=[title,status]
                status_id[bookid]="available"
                print("\n BOOK ID",bookid,"is added to the list \n")
            #when book selected is "Python"    
            elif Book_selected=="Python" and bookid not in Book_list["Python"]:
                Book_list["Python"][bookid]=[title,status]
                status_id[bookid]="available"
                print("\n BOOK ID",bookid,"is added to the list \n")
            #when book selected is ""Maths"
            elif Book_selected=="Maths" and bookid not in Book_list["Maths"]:
                Book_list["Maths"][bookid]=[title,status]
                status_id[bookid]="available"
                print("\n BOOK ID",bookid,"is added to the list \n")
            else:
                print("\n This Book Id is already present \n")#if the book id will wil already be present this message will be printed
            print("\n The list of books are: \n",Book_list)
        else:
            print("\n Please enter from the correct type of book \n") #if the type input is not from the given one
    elif user_menu==Menu[1]:
        while True:
            Book_selected=input('\n Please enter which type of book you want to delete from list \n Available options are "Fiction","Horror","Python","Maths": ' )
            print("Books in",Book_selected,"are: ",Book_list[Book_selected])
            if Book_selected=="Fiction" or Book_selected=="Horror" or Book_selected=="Python" or Book_selected=="Maths":
                bookid=int(input("\n Please enter the book Id you want to delete:"))
                #if input bookid is present in Fiction,Horror,Python or Maths availability will be checked
                if bookid in Book_list["Fiction"].keys() and status_id.keys() or bookid in Book_list["Horror"].keys() and status_id.keys() or bookid in Book_list["Python"].keys() and status_id.keys() or bookid in Book_list["Maths"].keys() and status_id.keys():
                    print("\n Checking for the availability status \n")
                    #checking if bookid is available or unavailable under status_id list
                    if status_id[bookid]=="available": #when the status shows as available
                        print("\n It's available \n")
                        #when book selected is fiction
                        if Book_selected=="Fiction":
                            del status_id[bookid] #the bookid will be deleted from the statusid dict
                            Book_list["Fiction"].pop(bookid) #the bookid dict will be deleted
                            print("\n The bookid",bookid,"is deleted \n")
                        #when book selected is "Horror"
                        elif Book_selected=="Horror":
                            del status_id[bookid]
                            Book_list["Horror"].pop(bookid)
                            print("\n The bookid",bookid,"is deleted \n")
                        #when book selected is "Python"
                        elif Book_selected=="Python":
                            del status_id[bookid]
                            Book_list["Python"].pop(bookid)
                            print("\n The bookid",bookid,"is deleted \n")
                        #when book selected is "Maths"    
                        elif Book_selected=="Maths":
                            del status_id[bookid]
                            Book_list["Maths"].pop(bookid) 
                            print("\n The bookid",bookid,"is deleted \n")
                            break
                        print("\n Books available after deletion are: \n",Book_list)
                        break
                    else:
                        print("\n It's issued, can't delete \n") #bookid is present but its status is unavialable in status_id dict
                        break
                else:
                    print("\n Invalid Id \n") #wrong id is input
            else:
                print("\n Please enter from the correct type of book \n")
    elif user_menu==Menu[2]: #when user input "View Book List"
        Book_selected=input('\n Please enter which type of book you want to View from list \n Available options are "Fiction","Horror","Python","Maths" :' )
        if Book_selected=="Fiction" or Book_selected=="Horror" or Book_selected=="Python" or Book_selected=="Maths":
            print("\n Books available in this category are(this includes issued books too):")
            #if the user input Fiction, the list of books under fiction will be printed
            if Book_selected=="Fiction":
                print(Book_list["Fiction"])
            elif Book_selected=="Horror":
                print(Book_list["Horror"])
            elif Book_selected=="Python":
                print(Book_list["Python"])
            elif Book_selected=="Maths":
                print(Book_list["Maths"])
            else:
                print("\n Please enter from the correct type of book \n")
        else:
            print("\n Please enter from the correct type of book \n")
    elif user_menu==Menu[3]: #when user input "Issue Book to Student"
        while True:
            print("\n The Books we have('this includes issued books')are: \n",Book_list)
            Book_selected=input('\n Please enter which type of book you want to issue from list \n Available options are "Fiction","Horror","Python","Maths" :')
            if Book_selected=="Fiction" or Book_selected=="Horror" or Book_selected=="Python" or Book_selected=="Maths":
                bookid=int(input("\n Please enter the book Id you want to issue:")) #input the book id user want to issue
                #if the book id is present in "Fiction" or "Horror" or "Python" or "Maths"
                if bookid in Book_list["Fiction"].keys() and status_id.keys() or bookid in Book_list["Horror"].keys() and status_id.keys() or bookid in Book_list["Python"].keys() and status_id.keys() or bookid in Book_list["Maths"].keys() and status_id.keys():
                #checking if input bookid is present ot not
                    print("\n Checking if its valid id \n")
                    print("\n This id is present \n")
                else:
                    print("\n Checking if its valid id \n")
                    print("\n This Book Id doesn't exist \n")
                    break
                #checking if bookid status is "available" or "unavailable" in status_id dict  
                if status_id[bookid]!="available":
                    print("\n THE BOOK IS ALREADY ISSUED \n") #if unavailabel, the book is issued
                issue_date=date.today() #this will store todays date
                return_date=issue_date+timedelta(days=15) #this will store the return date which is 15 days from issue date
                #storing the bookid, issue date and returndate in issue dict 
                if bookid in issue_dict.keys():
                    issue_dict[bookid].append([issue_date,return_date]) #if the bookid is not present in issue_dict it will get added as {bookid:[issuedate,returndate]}
                else:
                    issue_dict[bookid]=([issue_date,return_date]) #if bookid is present the status issue_dict will be shown as {bookid:[issuedate,returndate]}
                    status_id[bookid]="unavailable" #after issuig the status will chenge to unavailable in status_id dict
                    print("\n BOOK ID",bookid,"IS ISSUED")
                    print("\n You should return the book by",return_date)
                    break
            else:
                print("\n Please enter from the correct type of book \n")
    elif user_menu==Menu[4]:#If user enter Return Book
        while True:
            Book_selected=input('\n Please enter type of book you are returning \n Books are "Fiction","Horror","Python","Maths": ')
            if Book_selected=="Fiction" or Book_selected=="Horror" or Book_selected=="Python" or Book_selected=="Maths":
                bookid=int(input("\n Please enter the Book Id you want to return:"))
                #checking if the bookid exists in Booklist or status_id
                if bookid in Book_list["Fiction"].keys() and status_id.keys() or bookid in Book_list["Horror"].keys() and status_id.keys() or bookid in Book_list["Python"].keys() and status_id.keys() or bookid in Book_list["Maths"].keys() and status_id.keys():
                    print("\n Checking if its valid id")
                    print("\n This id is present")
                else:
                    print("\n Checking if its valid id")
                    print("\n This Book Id doesn't exist")
                    break
                if status_id[bookid]=="available": #if status of the book is "available" it means the book is not yet issued
                    print("\n This bookid was never issued \n")
                    break
                temp=issue_dict[bookid] #storing issue_dict value in temp
                today=date.today() #storing todays date
                #storing lenght of the issue_dict to temp_1
                for i in range(len(temp)):
                    temp_1=temp[i]
                return_date=temp_1 #storing the returndate from issue dict into return_date variable
                #when today date is greater than the return date, a fine will be charged
                if today>return_date:
                    print("\n The book is submitted {} days late. You have to pay Rs.10 as fine".format(today-return_date).days)
                else:
                    print("\n Thank You! for returning the book \n")
                issue_dict[bookid].remove(temp_1) #after returning the book removing it from issue_dict
                status_id[bookid]="available" #changing the status of bookid to available  
                print("\n Book Id",bookid,"is returned \n")
                break
            else:
                print("Please enter the correct type of book")
    elif user_menu==Menu[5]:#if user enters exit
        print("\n Thanks for visiting library. Have a good day :) ")
        user=False #after loop is over user need to login again
        break
    else: #if user enters anything other then the provided actions
        print("\n Sorry i dont understand the command")
```
