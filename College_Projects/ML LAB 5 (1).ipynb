### Importing necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### importing the breast cancer data

```python
data=pd.read_csv("C:\\Users\\A\\Desktop\\breast_cancer.csv")
data.head()
```

### separating the dependent and independent variables

```python
x=data.iloc[:,1:-1].values
y=data.iloc[:,-1].values
```

```python
x
```

### checking for null values

```python
data.isnull().sum()
```

### encoding the categorical dependent variabler

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
```

```python
print(y)
```

### splitting the dataset into train and test

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4,random_state=10)
```

### scaling all the independent variables

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

### fitting the logistic regression model with default parameters

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression(random_state=10)
classifier.fit(x_train,y_train)
```

### predicting the test set results

```python
y_pred=classifier.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

### accuary 

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

### checking change in accuracy for penalty='none'

```python
classifier1=LogisticRegression(penalty='none',solver="saga",random_state=10)
classifier1.fit(x_train,y_train)
```

```python
y_pred1=classifier.predict(x_test)
print(np.concatenate((y_pred1.reshape(len(y_pred1),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm1 = confusion_matrix(y_test, y_pred1)
print(cm1)
accuracy_score(y_test, y_pred1)
```

#### we can see that there is no change in accuracy of the model for penalty='none'

### model with penalty='l1'

```python
classifier2=LogisticRegression(penalty='l1',solver="saga",random_state=10)
classifier2.fit(x_train,y_train)
```

```python
y_pred2=classifier.predict(x_test)
print(np.concatenate((y_pred2.reshape(len(y_pred1),1), y_test.reshape(len(y_test),1)),1))
```

```python
cm2 = confusion_matrix(y_test, y_pred2)
print(cm2)
accuracy_score(y_test, y_pred2)
```
