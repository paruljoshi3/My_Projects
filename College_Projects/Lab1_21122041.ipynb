```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#1. Download the 80-cereals dataset from https://www.kaggle.com/crawford/80-cereals
df=pd.read_csv(r"D:\cereal.csv")
df
```

```python
#3. a. In the column "mfr", replace the column "K" as "Kellogg's", "G" as "Nestle" and all other values as "Other Brands"
df['mfr'].replace({'K':"Kellogg's",'G':"Nestle"},inplace=True)
df
```

```python
for i in df['mfr']:
    if i!="Kellogg's" and i!="Nestle":
        df['mfr'].replace({i:'Other Brands'},inplace=True)
```

```python
df
```

```python
#b. In the column "type", replace "C" with "Type 1" and "H" with "Type 2"
df['type'].replace({'C':"Type 1",'H':"Type 2"},inplace=True)
df
```

```python
#4. Visualise the count of above two features "mfr" and "type" with a bar-plot
sns.countplot(x='type',data=df)
```

```python
#8. Apply MinMaxScaler() and StandardScaler() to the following features: calories, protien, fat, sodium, fiber, carbo, sugars.
from sklearn.preprocessing import StandardScaler,MinMaxScaler,MaxAbsScaler
```

```python
scx=StandardScaler()
xs=scx.fit_transform(x)
xs
```

```python
x_1=df.iloc[:,3:10]
mmc=MinMaxScaler()
x_1=mmc.fit_transform(x_1)
x_1
```

```python
#1 more way for scaling is MaxAbs Scaler
x_2=df.iloc[:,3:10]
mas=MaxAbsScaler()
x_2=mas.fit_transform(x_2)
x_2
```

```python
#9. Does the standard or min-max scaling make a difference in value distribution? Support your answers with some visualisations on the above dataset.
plt.hist(xs,bins=10)
plt.title("Standard scaler")
```

```python
plt.hist(x_1,bins=10)
plt.title("MinMax scaler")
```

__Conclusion__:<br>
StandardScaler follows Standard Normal Distribution. Therefore, it makes mean = 0 and scales the data to unit variance. MinMaxScaler scales all the data features in the range [0, 1] or else in the range [-1, 1] if there are negative values in the dataset..
