```python
from sklearn.decomposition import PCA
from sklearn import preprocessing
from sklearn.model_selection import train_test_split as tts
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
scaler = preprocessing.StandardScaler().fit(X)
X_standardised = scaler.transform(X)
```

```python
X_standardised
```

```python
principal_component_analysis = PCA(n_components = 5)
results = principal_component_analysis.fit(X_standardised)
results_transformed = results.transform(X_standardised)
```

```python
results_transformed
```

```python
reduce_df = pd.DataFrame(data = results_transformed)
```

```python
reduce_df
```

```python
model_name=[]
dataset=[]
acc_score=[]
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
X_train, X_test, y_train, y_test = tts(df_X, df_y,random_state=42)

model.fit(X_train, y_train)
y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("The accuracy is : {}".format(accuracy))
print(classification_report(y_test, y_pred))
```

```python
y_pred
```

```python
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
model_name.append('DecisionTreeClassifier')
dataset.append('Entire dataset')
acc_score.append(accuracy)
```

```python
model1 = DecisionTreeClassifier()
X_train1, X_test1, y_train1, y_test1 = tts(reduce_df, df_y,random_state=42)

model1.fit(X_train1, y_train1)
y_pred1 = model1.predict(X_test1)

accuracy1 = accuracy_score(y_test1, y_pred1)
print("The accuracy is : {}".format(accuracy1))
print(classification_report(y_test1, y_pred1))
```

```python
y_pred1
```

```python
cm1=confusion_matrix(y_test1,y_pred1)
cm1
```

```python
model_name.append('DecisionTreeClassifier')
dataset.append('Using PCA')
acc_score.append(accuracy1)
```

```python
from sklearn import svm
```

```python
model2 = svm.SVC(kernel='linear')
X_train2, X_test2, y_train2, y_test2 = tts(df_X, df_y,random_state=42)

model2.fit(X_train2, y_train2)
y_pred2 = model2.predict(X_test2)

accuracy2 = accuracy_score(y_test2, y_pred2)
print("The accuracy is : {}".format(accuracy2))
print(classification_report(y_test2, y_pred2))
```
