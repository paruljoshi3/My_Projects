```python
#removing of outliers
from scipy.stats import zscore
z_score=abs(zscore(bs_new))
print(bs_new.shape)
bs_final=bs_new.loc[(z_score<3).all(axis=1)]
print(bs_final.shape)
```

```python
bs_final.skew()
```

```python
#removing the skewness
for index in bs_final.skew().index:
    if bs_final.skew().loc[index]>0.5:
        bs_final[index]=np.sqrt(bs_final[index])
```

```python
bs_final.skew()
```

```python
#assigning x and y values for training and testing
x=bs_final.drop(columns=["No. of predicted wins"])
y=bs_final[["No. of predicted wins"]]
```

```python
print(x)
print(y)
```

```python
x.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
def maxr2_score(regr,x,y):
    max_r_score=0
    for r_state in range(42,100):
        x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=r_state,test_size=0.22)
        regr.fit(x_train,y_train)
        y_pred=regr.predict(x_test)
        r2_scr=r2_score(y_test,y_pred)
        print("r2_score for :",r_state," is", r2_scr)
        if r2_scr>max_r_score:
            max_r_score=r2_scr
            final_r_state=r_state
    print("maximum r2_score to :",final_r_state," is :",max_r_score)
    return final_r_state
```

```python
lregr=LinearRegression()
```

```python
r_state=maxr2_score(lregr,x,y)
```

```python
#doing cross validation to check whether the model is suffering from overfitting or underfitting
from sklearn.model_selection import cross_val_score
cross_val_score(lregr,x,y,cv=5,scoring="r2").mean()
#not suffering from over fitting over under fitting
```

```python
#checking more models to see if any other model is scoring more than logisitic_Regression
#checking with decision_tree_classifier
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
```

```python
r_state=maxr2_score(dt,x,y)
```

```python
#doing cross validation to check whether the model is suffering from overfitting or underfitting
from sklearn.model_selection import cross_val_score
cross_val_score(dt,x,y,cv=5,scoring="r2").mean()
#not suffering from over fitting over under fitting
```

```python
#Using KNeighbors
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV

neighbors={"n_neighbors":range(1,20)}
knn=KNeighborsRegressor()
grs=GridSearchCV(knn,neighbors,cv=10)
grs.fit(x,y)
grs.best_params_
```

```python
knn=KNeighborsRegressor(n_neighbors=3)
r_state=maxr2_score(knn,x,y)
```

```python
from sklearn.linear_model import Lasso
lsreg=Lasso()
parameters={"alpha":[0.001,0.01,0.1,1]}
clf = GridSearchCV(lsreg, parameters, cv=10)
clf.fit(x,y)
clf.best_params_
```

```python
lsreg=Lasso(alpha=0.1)
r_state=maxr2_score(lsreg,x,y)
```

```python
#Lets use cross val score with Lasso
print("Mean r2 score for Lasso Regression: ",cross_val_score(lsreg,x,y,cv=5,scoring="r2").mean())
print("standard deviation in r2 score for Lasso Regression: ",cross_val_score(lsreg,x,y,cv=5,scoring="r2").std())
#Based on below output i can say Lasso perfored just like linear regression
```

```python
#Using ada boost regression algorithm
from sklearn.ensemble import AdaBoostRegressor
ada_reg=AdaBoostRegressor()
parameters={"learning_rate":[0.001,0.01,0.1,1],"n_estimators":[10,100,500,1000],"base_estimator":[lregr,dt,knn,lsreg]}
grs = GridSearchCV(ada_reg, parameters, cv=5)
grs.fit(x,y)
grs.best_params_
```

```python
ada_reg=AdaBoostRegressor(base_estimator=lsreg,learning_rate=1,n_estimators=10)
print("Mean r2 score for ada boosting Regression: ",cross_val_score(ada_reg,x,y,cv=5,scoring="r2").mean())
print("standard deviation in r2 score for ada boosting Regression: ",cross_val_score(ada_reg,x,y,cv=5,scoring="r2").std())
```

```python
r_state=maxr2_score(ada_reg,x,y)
```

```python
#As lasso is showing the best results so we will save it for the production
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=47,test_size=0.22)
lsreg=Lasso(alpha=0.1)
lsreg.fit(x_train,y_train)
y_pred=lsreg.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
print("r2 score is:" ,r2_score(y_test,y_pred))
print("RMSE is:" ,np.sqrt(mean_squared_error(y_test,y_pred)))
```
