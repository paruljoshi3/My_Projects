```python
from sklearn.preprocessing import LabelEncoder
```

```python
#lets seprate the input and output column
x=hr.drop(columns=["Attrition"])
y=hr[["Attrition"]]
```

```python
x=pd.get_dummies(x,drop_first=True)
```

```python
x
```

```python
#checking of skewness
x.skew()
```

```python
#removing the skewness for negative and positive both
for index in x.skew().index:
    if x.skew().loc[index]>0.5:
        x[index]=np.sqrt(x[index])
    if x.skew().loc[index]<-0.5:
        x[index]=np.sqrt(x[index])
```

```python
x.skew()
```

```python
le=LabelEncoder()
le.fit(y)
y=le.transform(y)
y
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(x)
x_s=sc.transform(x)
x_s=pd.DataFrame(x_s,columns=x.columns)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
from sklearn.metrics import roc_auc_score
```

```python
def max_aucroc(grs,df_x,y):
    max_aucroc=0
    for r_state in range(42,100):
        x_train, x_test, y_train, y_test = train_test_split(x, y,random_state = r_state,test_size=0.22)
        grs.fit(x_train,y_train)
        y_pred = grs.predict(x_test)
        aucroc_scr=roc_auc_score(y_test,y_pred)
        if aucroc_scr>max_aucroc:
            max_aucroc=aucroc_scr
            final_r_state=r_state
    print("max auc roc score for ",final_r_state," is ",max_aucroc)
    return final_r_state
```

```python
lregr=LogisticRegression()
```

```python
r_state=max_aucroc(lregr,x,y)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(lregr,x,y,cv=5,scoring="r2").mean()
#not suffering from over fitting over under fitting
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
r_state=max_aucroc(dt,x,y)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(dt,x,y,cv=5,scoring="r2").mean()
#not suffering from over fitting over under fitting
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
```

```python
neighbors={"n_neighbors":range(1,30)}
knn=KNeighborsClassifier()
grs=GridSearchCV(knn,neighbors,cv=10)
grs.fit(x,y)
grs.best_params_
```

```python
knn=KNeighborsClassifier(n_neighbors=14)
r_state=max_aucroc(knn,x,y)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(knn,x,y,cv=5,scoring="r2").mean()
#not suffering from over fitting over under fitting
```

```python
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.metrics import mean_squared_error
from sklearn.metrics import f1_score
```

```python
#As decision tree classifier is showing the best results so we will save it for the production
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=71,test_size=0.22)
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
y_pred=dt.predict(x_test)
```

```python
print("f1 score is:" ,f1_score(y_test,y_pred))
print("Confusion Matrix is:",confusion_matrix(y_test,y_pred))
print("Classification Report is:" ,classification_report(y_test,y_pred))
print("AUC ROC Score: ",roc_auc_score(y_test,y_pred))
```

```python
#Saving the model
from sklearn.externals import joblib
joblib.dump(dt,'HR_Analytics_dt.pk1')
```
